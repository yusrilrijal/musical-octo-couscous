```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

url = "https://man2-tulungagung.sch.id/data-alumni-man-2-tulungagung/"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

alumni_list = soup.find_all("tr")

data = []
for alumni in alumni_list:
    try:
        nama = alumni.find("td", class_="column-2").text
        jalur = alumni.find("td", class_="column-6").text.strip()  # Menghapus whitespace di sekitar teks
        data.append({
            "Nama": nama,
            "Jalur": jalur
        })
    except AttributeError as e:
        print("Error extracting data from alumni row:", e)

df = pd.DataFrame(data)

# Replace missing values in "Jalur" column with "noll"
df["Jalur"].fillna("noll", inplace=True)

print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

url = "https://man2-tulungagung.sch.id/data-alumni-man-2-tulungagung/"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

alumni_list = soup.find_all("tr")

data = []
for alumni in alumni_list:
    try:
        nama = alumni.find("td", class_="column-2").text
        jalur_elem = alumni.find("td", class_="column-6")
        jalur_text = jalur_elem.text.strip() if jalur_elem is not None else ""
        jalur = re.findall(r"\b\w+\b", jalur_text)
        jalur = jalur[0] if jalur else "noll"
        data.append({
            "Nama": nama,
            "Jalur": jalur
        })
    except AttributeError as e:
        print("Error extracting data from alumni row:", e)

df = pd.DataFrame(data)

print(df)
```
